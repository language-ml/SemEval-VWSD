```python
def get_img_emb(name, image, model):
    if name=='clip':
        image = preprocess(image)
        image_input = torch.tensor(np.stack(image)).cuda()
        image_input = image_input[None, :, :, :]
        with torch.no_grad():
            image_features = model.encode_image(image_input).float()
        image_features /= image_features.norm(dim=-1, keepdim=True)
        return image_features
    
def get_text_emb(name, text, model):
    if name=='clip':
        text_tokens = clip.tokenize(text, truncate=True).cuda()
        with torch.no_grad():
            text_features = model.encode_text(text_tokens).float()
        text_features /= text_features.norm(dim=-1, keepdim=True)
        return text_features

def load_finetuned_model_muwosed(model_path):
    clip_text_encoder = finetuner.get_model(model_path, select_model='clip-text')
    clip_image_encoder = finetuner.get_model(model_path, select_model='clip-vision')
    return clip_text_encoder, clip_image_encoder

def get_img_emb_muwosed(img, model):
    doc = DocumentArray([Document(tensor=img)])
    img_ft = finetuner.encode(model=model, data=doc)
    return img_ft.embeddings
    
def get_text_emb_muwosed(text, model):
    text_da = DocumentArray([Document(text=text)])
    txt_ft = finetuner.encode(model=model, data=text_da)
    return txt_ft.embeddings
```

```python
import io

class CPU_Unpickler(pickle.Unpickler):
    def find_class(self, module, name):
        if module == 'torch.storage' and name == '_load_from_bytes':
            return lambda b: torch.load(io.BytesIO(b), map_location='cpu')
        else: return super().find_class(module, name)
```

# Run1(Finetune on train)

```python
class CustomSemEvalDataset(Dataset):
    def __init__(self, data, label, mode, muwosed=False):
        self.data = data
        self.label = label
        self.mode = mode
        self.muwosed = muwosed

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data[idx]
        if not self.muwosed:
            if self.mode=='train':
                text_emb = text_train_emb_clip[row[1]]
                img_emb = []
                for i in row[2:]:
                    img_emb.append(img_train_emb_clip[i].squeeze())
            elif self.mode=='test':
                text_emb = text_test_emb_clip[row[1]].cpu()
                img_emb = []
                for i in row[2:]:
                    img_emb.append(img_test_emb_clip[i].squeeze().cpu())
        else:
            if self.mode=='train':
                text_emb = text_train_emb_muwosed[row[1]]
                img_emb = []
                for i in row[2:]:
                    img_emb.append(img_train_emb_muwosed[i].squeeze())
            elif self.mode=='test':
                text_emb = text_test_emb_muwosed[row[1]]
                img_emb = []
                for i in row[2:]:
                    img_emb.append(img_test_emb_muwosed[i].squeeze())
        img_emb = np.stack(img_emb)

        label_idx = np.where(self.label[idx]==row[2:])[0][0]
        label = np.zeros(10)
        label[label_idx] = 1
        return img_emb, text_emb.squeeze(), label, label_idx, list(row[2:])

from torch.utils.data import DataLoader

train_data = CustomSemEvalDataset(train_file, train_gold, 'train')
test_data = CustomSemEvalDataset(test_file, test_gold, 'test')

train_dataloader = DataLoader(train_data, batch_size=8, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=8)
```

```python
class Finetune(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1_img = nn.Linear(512, 512)
        # self.fc2_img = nn.Linear(1024, 512)
        self.fc1_text = nn.Linear(512, 512)
        # self.fc2_text = nn.Linear(1024, 512)

    def forward(self, img, text):
        # img = F.relu(self.fc1_img(img))
        # img = self.fc2_img(img)
        img = self.fc1_img(img)
        # text = F.relu(self.fc1_text(text))
        # text = self.fc2_text(text)
        text = self.fc1_text(text)
        text = torch.nn.functional.normalize(text, dim=-1)
        img = torch.nn.functional.normalize(img, dim=-1)
        return img, text

class Finetune2(nn.Module):
    def __init__(self):
        super().__init__()
        # self.fc1_img = nn.Linear(512, 256)
        # self.fc2_img = nn.Linear(256, 256)
        # self.fc1_text = nn.Linear(512, 256)
        # self.fc2_text = nn.Linear(256, 256)

    def forward(self, img, text):
        # img = F.relu(self.fc1_img(img))
        # img = self.fc2_img(img)
        # text = F.relu(self.fc1_text(text))
        # text = self.fc2_text(text)
        return img, text
chert = Finetune2()
chert.to(device)
```

```python
EPOCH = 10

finetune = Finetune()
finetune.to(device)
loss = nn.CrossEntropyLoss()
optimizer = optim.Adam(finetune.parameters(), lr=1e-3)

for epoch in range(EPOCH):
    print('Epoch {}/{}'.format(epoch, epoch-1))
    print('-' * 10)
    running_loss = 0.0
    for batch in tqdm(train_dataloader):

        images, text, label, _, _ = batch
        
        optimizer.zero_grad()
        images = torch.Tensor(images)
        images = images.to(device)
        text = text.to(device)
        label = label.to(device)
        
        logits_image, logits_text = finetune(images, text)
        sim = torch.einsum('Bi,BNi->BN', logits_text, logits_image)

        total_loss = loss(sim, label)
        total_loss.backward()
        optimizer.step()

        running_loss += total_loss.item() * images.size(0)

    epoch_loss = running_loss / len(train_dataloader)
    print('Loss: {:.4f}'.format(epoch_loss))
```

```python
hit = 0
mrr = 0

with torch.no_grad():
    for batch in tqdm(test_dataloader):
        images, text, label, label_idx, _ = batch
    
        images = torch.Tensor(images)
        images = images.to(device)
        text = text.to(device)
        label = label.to(device)
        
        logits_image, logits_text = finetune(images, text)
        sim = torch.einsum('Bi,BNi->BN', logits_text, logits_image).cpu()

        hit +=((np.array(np.argmax(sim, axis=1))==np.array(label_idx)) == True).sum()
        for i, j in zip(sim, label_idx):
            mrr+=1/(np.where(np.argsort(-i) == int(j))[0]+1)


print(hit/len(test_file))
print(mrr/len(test_file))
```
