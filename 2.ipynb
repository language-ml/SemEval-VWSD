```python
from pkg_resources import packaging
from PIL import Image, ImageFile
from tqdm import tqdm
from torch import nn
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets, models
from sklearn import metrics
import torch.nn.functional as F
import torch.optim as optim
import os
import pickle
import numpy as np
import torch
import skimage
import IPython.display
import matplotlib.pyplot as plt


%matplotlib inline

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

Image.MAX_IMAGE_PIXELS = None
ImageFile.LOAD_TRUNCATED_IMAGES = True

print("Torch version:", torch.__version__)
```

```python
import clip

clip.available_models()
```

```python
model_clip, preprocess = clip.load("ViT-B/16", device=device, jit=False)

input_resolution = model_clip.visual.input_resolution
context_length = model_clip.context_length
vocab_size = model_clip.vocab_size

print("Model parameters:", f"{np.sum([int(np.prod(p.shape)) for p in model_clip.parameters()]):,}")
print("Input resolution:", input_resolution)
print("Context length:", context_length)
print("Vocab size:", vocab_size)
```

```python
base_train_dir = 'semeval-2023-task-1-V-WSD-train-v1/train_v1/train_images_v1/'

with open('semeval-2023-task-1-V-WSD-train-v1/train_v1/train.data.v1.txt') as f:
    train_file = f.readlines()
train_file = np.array([line.strip().split('\t') for line in train_file])

all_train_img = set(train_file[:, 2:].flatten())
all_train_text = set(train_file[:, 1].flatten())

with open('semeval-2023-task-1-V-WSD-train-v1/train_v1/train.gold.v1.txt') as f:
    train_gold = f.read()
train_gold = train_gold.split("\n")[:-1]
```

```python
base_test_dir = 'test_images/'

with open('en.test.data.v1.1.txt') as f:
    test_file = f.readlines()
test_file = np.array([line.strip().split('\t') for line in test_file])

all_test_img = set(test_file[:, 2:].flatten())

with open('en.test.gold.v1.1.txt') as f:
    test_gold = f.read()
test_gold = test_gold.split("\n")[:-1]

with open('prompts.txt') as f:
    all_test_text = f.read()

all_test_text = all_test_text.split("\n")

test_file = test_file.astype('<U500')
for i, j in zip(test_file, all_test_text):
    i[1] = j
```

```python
class image_title_dataset(Dataset):
    def __init__(self, list_image_path, list_txt):

        self.image_path = list_image_path
        self.txt = clip.tokenize(list_txt)

    def __len__(self):
        return len(self.txt)

    def __getitem__(self, idx):
        image = preprocess(Image.open(os.path.join(base_train_dir,self.image_path[idx])).convert("RGB"))
        txt = self.txt[idx]
        return image, txt

dataset = image_title_dataset(train_gold, train_file[:,1])
train_dataloader = DataLoader(dataset, batch_size=8)
```

```python
def convert_models_to_fp32(model): 
    for p in model.parameters(): 
        p.data = p.data.float() 
        p.grad.data = p.grad.data.float() 


if device == "cpu":
  model_clip.float()
else :
  clip.model.convert_weights(model_clip)

loss_img = nn.CrossEntropyLoss()
loss_txt = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_clip.parameters(), lr=1e-6)


for epoch in tqdm(range(5)):
    running_loss = 0.0
    for batch in tqdm(train_dataloader):
        optimizer.zero_grad()

        images,texts = batch 
        
        images= images.to(device)
        texts = texts.to(device)
        
        logits_per_image, logits_per_text = model_clip(images, texts)

        ground_truth = torch.arange(len(images),dtype=torch.long,device=device)

        total_loss = (loss_img(logits_per_image,ground_truth) + loss_txt(logits_per_text,ground_truth))/2
        total_loss.backward()
        if device == "cpu":
            optimizer.step()
        else : 
            convert_models_to_fp32(model_clip)
            optimizer.step()
            clip.model.convert_weights(model_clip)

    running_loss += total_loss.item() * images.size(0)
    epoch_loss = running_loss / len(train_dataloader)
    print('Loss: {:.4f}'.format(epoch_loss))
```

```python
class test_dataset(Dataset):
    def __init__(self, test_file, label):
        self.test_file = test_file
        self.label = label

    def __len__(self):
        return len(self.test_file)

    def __getitem__(self, idx):
        row = self.test_file[idx]

        text_tokens = clip.tokenize(row[1], truncate=True).cuda()
        with torch.no_grad():
            text_features = model_clip.encode_text(text_tokens).float()
        text_features /= text_features.norm(dim=-1, keepdim=True)
        img_emb = []
        for i in row[2:]:
            image = Image.open(os.path.join(base_test_dir, i)).convert("RGB")
            image = preprocess(image)
            image_input = torch.tensor(np.stack(image)).cuda()
            image_input = image_input[None, :, :, :]
            with torch.no_grad():
                image_features = model_clip.encode_image(image_input).float()
            image_features /= image_features.norm(dim=-1, keepdim=True)
            img_emb.append(image_features.squeeze().cpu())

        img_emb = np.stack(img_emb)
        label_idx = np.where(self.label[idx]==row[2:])[0][0]
        return img_emb, text_features.squeeze(), label_idx

dataset = test_dataset(test_file, test_gold)
test_dataloader = DataLoader(dataset, batch_size=8)

hit = 0
mrr = 0

with torch.no_grad():
    for batch in tqdm(test_dataloader):
        images, text, label_idx = batch
    
        images = torch.Tensor(images)
        images = images.to(device)
        text = text.to(device)
        # label_idx = label_idx.to(device)

        sim = torch.einsum('Bi,BNi->BN', text, images).cpu()

        hit +=((np.array(np.argmax(sim, axis=1))==np.array(label_idx)) == True).sum()
        for i, j in zip(sim, label_idx):
            mrr+=1/(np.where(np.argsort(-i) == int(j))[0]+1)


print(hit/len(test_file))
print(mrr/len(test_file))
```

```python
pairs = DocumentArray() # initialize a DocumentArray as final training data.

prompt = 'This is a photo of '
for i,j in tqdm(zip(train_file[:100], train_gold[:100])):
    if j=='image.12060.jpg' or j=='image.12229.jpg' or j=='image.3597.jpg':
        continue
    pair = Document()
    doc = Document(uri=f"{base_train_dir}{j}")
    try:
        img_chunk = doc.load_uri_to_image_tensor(224, 224)
    except:
        continue
    img_chunk.modality = 'image'
    txt_chunk = Document(content=prompt + i[1])
    txt_chunk.modality = 'text'
    pair.chunks.extend([img_chunk, txt_chunk])
    # add pair to pairs
    pairs.append(pair)

# # Lets see the first item of the pairs
pairs[0]
```

```python
from finetuner.callback import WiSEFTCallback
def fune_tune(model, name):
    run = finetuner.fit(
        model=model,
        # experiment_name='clip-finetune-exp-1',
        run_name=name,
        train_data='train-data-clip',
        epochs=5,
        learning_rate=1e-7,
        loss='CLIPLoss',
        device='cuda',
        callbacks=[
            WiSEFTCallback(alpha=0.5)
        ],
        freeze=False,
    )
    for entry in run.stream_logs():
        print(entry)
    artifact = run.save_artifact('./Models/')
```

```python
fune_tune('ViT-B-32::openai', 'vit32_finetune_on_train_systempaper')
```

```python
def load_finetuned_model(model_path):
    clip_text_encoder = finetuner.get_model(model_path, select_model='clip-text')
    clip_image_encoder = finetuner.get_model(model_path, select_model='clip-vision')
    return clip_text_encoder, clip_image_encoder

def get_img_emb(img, model):
    doc = DocumentArray([Document(tensor=img)])
    img_ft = finetuner.encode(model=model, data=doc)
    return img_ft.embeddings
    
def get_text_emb(text, model):
    text_da = DocumentArray([Document(text=text)])
    txt_ft = finetuner.encode(model=model, data=text_da)
    return txt_ft.embeddings
```

```python
text_encoder, image_encoder = load_finetuned_model('drive/MyDrive/SemEval1/VWSD/ACL/Models/vit32_finetune_on_train_systempaper.zip')
```

```python
img_test_emb = {}

for i in tqdm(all_test_img):
    img = np.array(Image.open(os.path.join(base_test_dir, i)).convert("RGB"))
    img_test_emb[i] = get_img_emb(img, image_encoder)

text_test_emb = {}

for i in tqdm(all_test_text):
    text_test_emb[i] = get_text_emb(i, text_encoder)
```

```python
class CustomSemEvalDataset(Dataset):
    def __init__(self, data, label):
        self.data = data
        self.label = label

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data[idx]
        
        text_emb = text_test_emb[row[1]]
        img_emb = []
        for i in row[2:]:
            img_emb.append(img_test_emb[i].squeeze())
        img_emb = np.stack(img_emb)

        label_idx = np.where(self.label[idx]==row[2:])[0][0]
        # label = np.zeros(10)
        # label[label_idx] = 1
        return (text_emb.squeeze(), img_emb), label_idx

from torch.utils.data import DataLoader

test_data = CustomSemEvalDataset(test_file, test_gold)

test_dataloader = DataLoader(test_data, batch_size=8)
```

```python

hit = 0
mrr = 0

with torch.no_grad():
    for data, label in test_dataloader:
        data[0] = data[0].squeeze()
        data[1] = data[1].squeeze()
        sim = torch.einsum('Bi,BNi->BN', data[0], data[1]).cpu()
        hit +=((np.array(np.argmax(sim, axis=1))==np.array(label)) == True).sum()
        for i, j in zip(sim, label):
            mrr+=1/(np.where(np.argsort(-i) == int(j))[0]+1)


print(hit/len(test_file))
print(mrr/len(test_file))
```
